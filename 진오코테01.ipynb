## Q5, 6

```python
def choices(weights, k=1):
    result = ()
    count = 0
    while count < k:
        target = sum(weights) * random.random()
        temp = 0
        for i in range(len(weights)):
            temp += weights[i]
            if temp >= target:
                if result.count(i) == 0:
                    result += (i, )
                    count += 1
                break
    return result
```

```python
weights = [2**x for x in range(7, -1, -1)]
```

```python
weights
```

```python
a = []
for n in range(10000):
    a.append(choices(weights))
    
a = pandas.Series(a)
round(a.value_counts()/10000, 2)
```

```python
a = []
for n in range(10000):
    a.append(choices(weights, k=2))
    
a = pandas.Series(a)
round(a.value_counts()/10000, 2)
```
